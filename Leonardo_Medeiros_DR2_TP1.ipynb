### Exercício 11 – Otimização de Carga e Estabilidade de Frete
Um operador logístico trabalha com o transporte de insumos industriais em caminhões de pequeno porte. Cada saca de polímero pesa exatamente 65kg, e o veículo possui um limite de carga útil de 550kg. O impacto esperado é garantir que o caminhão nunca ultrapasse sua capacidade máxima, transportando apenas unidades inteiras para evitar o deslocamento perigoso da carga durante o trajeto. 

**Tarefa:**
Com base na capacidade de 550kg e no peso unitário de 65kg, desenvolva um código que calcule o número máximo de sacas que podem ser embarcadas em um caminhão sem exceder o limite, bem como a porcentagem de ocupação do peso do caminhão . O resultado deve ser armazenado em uma variável que represente a capacidade real, e outra com a porcentagem, ambas exibidas no console.

```
saca_polimero = 65
limite_carga = 550
num_max = limite_carga // saca_polimero
capacidade = num_max * saca_polimero
porcentagem = (capacidade / limite_carga) * 100

print(f'Podem ser embarcadas no caminhão {num_max} sacas.')
print(f'Essas sacas ocupam um total de {porcentagem:.2f}% do peso do caminhão')
```

### Exercício 15 – Telemetria Aeroespacial e Autonomia de Voo
Um drone de monitoramento ambiental realizou dois voos de mapeamento um com duração de 3.77 horas enquanto o outro foram 214 minutos. No entanto, o sistema de telemetria de bateria processa o consumo de energia exclusivamente em segundos. Para calcular a energia total que a bateria deve ter, considere que cada segundo do drone corresponde à 2,7 joules. O impacto esperado é o cálculo preciso da energia necessária para evitar a perda do equipamento durante o pouso, adicionando 15% de margem de segurança.

**Tarefa:**
A partir do tempo de ambos os voos, desenvolva o raciocínio lógico para converter o total integralmente para a unidade de segundos, e obter quanto de energia foi gasto nos voos. Em seguida, considere a margem de segurança e armazene o resultado final em uma variável apropriada e exiba no terminal.

```
joules = 2.7
margem = 0.15 
tempo_voo1 = 3.77
tempo_voo2 = 214

segundos_voo1 = tempo_voo1 * 3600
segundos_voo2 = tempo_voo2 * 60
total_segundos = segundos_voo1 + segundos_voo2

energia_voo1 = segundos_voo1 * joules
energia_voo2 = segundos_voo2 * joules

energia = joules * total_segundos
energia_total = energia * (1 + margem)

print(f'Tempo do primeiro voo: {segundos_voo1:.0f} segundos\n') 
print(f'Tempo do segundo voo: {segundos_voo2:.0f} segundos\n')
print(f'Tempo total do voo: {total_segundos:.0f} segundos\n') 
print(f'Energia total: {energia_total:.2f} Joules')  
```

### Exercício 16 – Decomposição de Jornada para Fechamento de Folha

Um técnico de som trabalhou em um set de filmagem por um total de 1527 minutos. Para o fechamento do contrato, o RH precisa que o tempo seja discriminado separando as horas completas dos minutos excedentes, pois as horas e os minutos possuem valores de pagamento distintos. O impacto esperado é a transparência e conformidade com os acordos sindicais. Para este contrato cada hora completa trabalhada é paga a R$ 60,00 e cada minuto excedente é pago a R$ 1,20. O impacto esperado é conseguir definir com precisão o valor que deverá ser pago.

**Tarefa:** A partir do total de 1527 minutos, calcule a quantidade de horas inteiras trabalhadas e a quantidade de minutos restantes. Em seguida calcule o valor total referente às horas e o valor total referente aos minutos excedentes. 

Exiba no terminal devidamente sinalizada:
   - horas trabalhadas  
   - minutos excedentes  
   - valor pago pelas horas  
   - valor pago pelos minutos  
   - valor total a receber  

```
hora_completa = 60 
minutos_excedente = 1.20 

tempo_total = 1527
hora = tempo_total // 60
minutos = tempo_total % 60

valor_horas = hora * hora_completa
valor_minutos = minutos * minutos_excedente
valor_total = valor_horas + valor_minutos

print(f'Horas trabalhadas: {hora} horas\n')
print(f'Minutos excedentes: {minutos} minutos\n')
print(f'Valor pago pelas horas: R$ {valor_horas:.2f}\n')
print(f'Valor pago pelos minutos: R$ {valor_minutos:.2f}\n')
print(f'Valor total a receber: R$ {valor_total:.2f}')

```
